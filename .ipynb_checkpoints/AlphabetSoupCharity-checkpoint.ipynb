### Deliverable 1: Preprocessing the Data for a Neural Network

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#clean_application_df = application_df.drop('EIN', 1)
#clean_application_df = clean_application_df.drop('NAME', 1)

clean_application_df = application_df.drop(['EIN', 'NAME'], 1)

clean_application_df.head()
```

```python
# Determine the number of unique values in each column.
charity_list = clean_application_df.dtypes[clean_application_df.dtypes == 'object'].index.tolist()
clean_application_df[charity_list].nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
Binning_Series = clean_application_df['APPLICATION_TYPE'].value_counts()
Binning_Series
```

```python
# Visualize the value counts of APPLICATION_TYPE

Binning_Series.plot.density()
```

```python
# Determine which values to replace if counts are less than ...?

Clean_Binning_Series = Binning_Series[Binning_Series < 200].index.tolist()

# Replace in dataframe
for app_type in Clean_Binning_Series:
    clean_application_df['APPLICATION_TYPE'] = clean_application_df['APPLICATION_TYPE'].replace(app_type, 'Others')

clean_application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
CLASS_Series = clean_application_df['CLASSIFICATION'].value_counts()
CLASS_Series
```

```python
# Visualize the value counts of CLASSIFICATION
CLASS_Series.plot.density()
```

```python
# Determine which values to replace if counts are less than ..?

clean_CLASS_list = CLASS_Series[CLASS_Series < 1800].index.tolist()

for class_type in CLASS_Series:
     clean_application_df['CLASSIFICATION'] = clean_application_df['CLASSIFICATION'].replace(class_type, 'Others')

clean_application_df['CLASSIFICATION'].value_counts()
```

```python
# Generate our categorical variable lists
clean_application_df[charity_list].nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(clean_application_df[charity_list]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(charity_list)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
encoded_charity_df  = clean_application_df.merge(encode_df, left_index = True, right_index = True)\
                                .drop(columns = charity_list)
encoded_charity_df.head()
```

```python
# Split our preprocessed data into our features and target arrays

y = encoded_charity_df['IS_SUCCESSFUL'].values
X = encoded_charity_df.drop(columns = ['IS_SUCCESSFUL']).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Deliverable 2: Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
num_input = len(X_train_scaled[0])
num_first = len(X_train_scaled[0])*2
num_second = 50



nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=num_first, input_dim = num_input,
                                    activation ='relu'))

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units = num_second, activation='relu'))

# Output layer
nn_model.add(tf.keras.layers.Dense(units = num_second, activation='relu'))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss = 'binary_crossentropy', optimizer = 'adam',
                metrics = ['accuracy'])
```

```python
# Train the model
model_history = nn_model.fit(X_train_scaled, y_train, epochs=200)

history_df = pd.DataFrame(model_history.history, 
                    index = range(1, len(model_history.history['loss'])+1))
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
